```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.random.random(1000)*100
Y_data = np.random.random(1000)*100
# the random function generates numbers bw 0 and 1. so when u do *100 itll be bw 0 and 100
```

**Scatter Plots**

Its a bunch of data points plotted as dots

```python
# c: colour, s: size of the points, alpha is used to provide transperancy (helpful when youre working with more data points)
plt.scatter(X_data, Y_data,c="purple",marker="*",s=100,alpha=0.3)
plt.show()
```

**Line Plots**

```python
# lets say we wanna analyse weight change of a person over few years
years = [2006+x for x in range(16)]
weights = [80,83,84,85,86,83,80,78,81,79,78,80,80,82,79,83]
plt.plot(years, weights,c="g",lw=2,linestyle="--")
# eqv: plt.plot(years, weights,"r--",lw=2)
# here you dont have to do plt.line because plot() by default plots a line chart
plt.show()
```

**Bar Plots**

```python
# lets say theres a poll on which programming language people like and you wanna analyse data
x = ["c++","c#","python","java","go"]
y = [20,50,140,10,45]

plt.bar(x,y,color="pink",align="edge", width=0.5,edgecolor="blue",lw=2)
# align basically positions the text on the x axis wrt to the bar
```

**Histogram**

```python
# A normal distribution is a symmetric, bell-shaped curve where most values cluster around the mean, and the probability of values decreases as you move further from the mean
ages = np.random.normal(20,1.5,1000)
# so here we are generating 1000 random numbers centered arond the mean 20 with a std dev of 1.5
plt.hist(ages,bins=20,cumulative=True)
# plt.hist(ages,bins=[ages.min(),18,21,ages.max()])
```

**Pie Charts**

```python
# who uses what programming language most frequently (non overlapping)
langs=["Python","C++","Java","C#","Go"]
votes=[50,24,14,6,17]
explodes=[0,0,0.2,0,0]
plt.pie(votes,labels=langs,explode=explodes,autopct="%.2f%%",pctdistance=0.6,startangle=90)
# explode is used if u want to pull out a segment to highlight it more
# autopct is to show the percentages, pctdistance tells how far from the center to place the %
```

**Box Plots**

It shows different quartiles,medians, max, min and stuff

```python
heights=np.random.normal(172,8,300)
plt.boxplot(heights)
plt.show()
# how to read a box plot

""" the box basically represents the middle 50% of your data
which goes from 25th to 75th percentile.
the middle line represents the median which tells that 50% of the ppl are shorter than that height
lines extending from the box called whiskers basically just show the spread of your other 50% of the data
horizontal lines at the end of the whiskers represent max and min of normal data
outliers (circles outside the normal data) represent unusually tall or short people"""
```

```python
years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
income = [55,56,62,61,72,72,73,75]
income_ticks = list(range(50,81,2))
plt.plot(years,income)
plt.title("Income of John (in USD)",fontsize=15)
plt.xlabel("Year")
plt.ylabel("Yearly Income (in USD)")
plt.yticks(income_ticks,[f"{x}k" for x in income_ticks])
plt.show()
```

**Multiple Figures**

```python
x1,y1=np.random.random(100),np.random.random(100)
x2,y2=np.arange(100), np.random.random(100)
plt.figure(1)
plt.scatter(x1,y1)

plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

```python
x=np.arange(100)
fig,axs=plt.subplots(2,2) # plotting a 2*2 grid in one figure so you can have four different plots

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine Wave")

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Cosin Wave")

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title("Random Function")

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title("Log Function")
axs[1,1].set_xlabel("TEST")

fig.suptitle("Four Plots")
plt.tight_layout() # avoids overlap between the subplots
plt.show()

plt.savefig("fourplots.png",dpi=300, transparent=False) # exports the plot

```

**3D Plotting**

```python
ax=plt.axes(projection="3d")
x=np.random.random(100)
y=np.random.random(100)
z=np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Plot")
ax.set_xlabel("X-axis")
plt.show()
```

```python
ax=plt.axes(projection="3d")
x=np.arange(0,50,0.1)
y=np.sin(x)
z=np.cos(x)

ax.plot(x,y,z)
```

Surface Plots

```python
ax=plt.axes(projection="3d")
x=np.arange(-5,5,0.1)
y=np.arange(-5,5,0.1)
# x and y are 1D arrays and now we're going to create a grid out of it
X, Y = np.meshgrid(x,y)
Z=np.sin(X)+np.cos(Y)

ax.plot_surface(X,Y,Z,cmap="Spectral")
ax.view_init(azim=45,elev=45) # azim controls right and left view. elev controls top and bottom view
```
